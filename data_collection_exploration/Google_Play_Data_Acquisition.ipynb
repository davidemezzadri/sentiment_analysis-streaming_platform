```python
import pandas as pd
import numpy as np
from google_play_scraper import app, Sort, reviews_all, reviews
```

## Scraping of Netflix reviews and saving to a CSV

```python
netflix, continuation_token_netflix = reviews('com.netflix.mediaclient', lang = 'en', country = 'US', sort = Sort.NEWEST, count = 1000000)
```

```python
netflix = pd.json_normalize(netflix)
```

```python
netflix
```

```python
netflix.to_csv('Netflix_Google_Play.csv')
```
